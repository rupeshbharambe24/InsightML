# Model Battle

Run parallel cross-validation across multiple models and compare results.

```python
import dissectml as dml

df = dml.load_titanic()
```

## Run the Battle

```python
models = dml.battle(
    df,
    target="survived",
    task="classification",
    families=["linear", "tree", "ensemble"],
    cv=5,
    n_jobs=-1,
)
```

## Leaderboard

```python
models.leaderboard()
```

## Model Comparison

```python
comp = dml.ModelComparator(models, y=df["survived"])
comp.table.styled()
```

```python
comp.pareto.show()
```

## Statistical Significance

```python
sig = comp.significance
sig["mcnemar"]["figure"].show()
```

## Error Analysis

```python
ea = comp.error_analysis
ea.disagreement_figure().show()
```

## ROC Curves

```python
comp.roc_curves.show()
```

## Hyperparameter Tuning

```python
tuned = dml.battle(
    df, target="survived",
    families=["tree"],
    tune=True, top_n=2, n_iter=10,
)
tuned.leaderboard()
```
